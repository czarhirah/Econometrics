# Training and Testing MVLR

```python
# Splitting the data into features (X) and target variable (y)
X = fulldata[['LFD Policy Rate', 'LFD M2', 'L GDP', 'FD CA', 'US Fed Interest Rate']]
y = fulldata['FX Rate'].values.flatten()  # Flattening y to ensure it's a 1D array

# Splitting data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Training the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predicting FX rate for the test set
y_pred = model.predict(X_test)

# Plotting the multivariate regression with legends, variable annotations, and R-squared value
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', label='Actual vs. Predicted')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Perfect Prediction')

plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Multivariate Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

# Calculating RMSE
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

# Calculate RMSE
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))
rmse_value = rmse(y_test, y_pred)

def rmspe(y_true, y_pred):
    percentage_errors = ((y_true - y_pred) / y_true) ** 2
    mean_squared_percentage_error = np.mean(percentage_errors)
    rmspe = np.sqrt(mean_squared_percentage_error)
    return rmspe

# Calculate RMSPE
rmspe_value = rmspe(y_test, y_pred)

# Extracting dates for actual and predicted comparison plot
xdates = X_test.reset_index()
xdates.rename(columns={'index': 'Date'}, inplace=True)
xdates = xdates['Date']

# Convert Series to DataFrame
validationdata = pd.DataFrame(xdates, columns=['Date'])
validationdata['y_test'] = y_test
validationdata['y_pred'] = y_pred

# Convert 'Date' to datetime
validationdata['Date'] = pd.to_datetime(validationdata['Date'], format='%b-%y')

# Sort by 'Date'
validationdata = validationdata.sort_values(by='Date')

# Reset index after sorting
validationdata = validationdata.reset_index(drop=True)

# Plotting the actual and predicted FX rates with dates
plt.figure(figsize=(10, 6))
plt.plot(validationdata['Date'],validationdata['y_test'], color='blue', label='Actual FX Rate')
plt.plot(validationdata['Date'],validationdata['y_pred'],color='red', linestyle='--', label='Predicted FX Rate')

plt.xlabel('Date')
plt.ylabel('FX Rate')
plt.title('Actual vs. Predicted FX Rates')
plt.legend()
plt.xticks(rotation=45) 
#label_interval = 3  # Showing every 5th label
#plt.xticks(ticks=fulldata.index[::label_interval]) 
plt.grid(True)
plt.show()

# Calculate the mean of the observations
mean_y_test = np.mean(y_test)

# Calculate total sum of squares
ss_tot = np.sum((y_test - mean_y_test) ** 2)

# Calculate residual sum of squares
ss_res = np.sum((y_test - y_pred) ** 2)

# Calculate R-squared
r_squared_manual = 1 - (ss_res / ss_tot)

# Calculate RMSE
rmsedata = rmse(y_test, y_pred)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate Mean Absolute Percentage Error (MAPE)
def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
mape = mean_absolute_percentage_error(y_test, y_pred)

# Print the metrics
print("-" * 50)

print(f"MSE: {mse}")
print(f"RMSE: {rmsedata}")
print(f"RMSPE: {rmspe_value}")
print(f"R-squared: {r_squared_manual:.2f}")
```

def adf_check(data, variable_name): 
    adftest = adfuller(data[variable_name], autolag='AIC', regression='ct')
    print("ADF Test Result - "+variable_name)
    print("Null Hypothesis: The series has a unit root (non-stationary)")
    print("ADF-Statistic:", adftest[0])
    print("P-Value:", adftest[1])
    print("Number of lags:", adftest[2])
    print("Number of observations:", adftest[3])
    print("Critical Values:", adftest[4])
    if adftest[1] <= 0.1:
        print("\033[1mThe series is stationary (P-Value <= 0.1)\033[0m")
        print("\n")
        return 0
    else:
        print("\033[1mThe series is not stationary (P-Value > 0.1)\033[0m")
        print("\n")
        return 1

        adf_results_before.loc[adf_results_before['Variable'] == variable_name, 'ADF Test Result'] = result_before

        # Perform ADF test after transformation
        result_after = adf_check(ivdata, 'Diff ' + variable_name if transformed else variable_name)

        if transformed:
